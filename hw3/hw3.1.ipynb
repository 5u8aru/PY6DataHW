```python
import numpy as np

import pandas as pd

tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C", match = 'Коефіцієнт народжуваності в регіонах України', thousands='.', decimal=',')
table = tmp[0]
table
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
years = table['1960'][:27]
region = table['Регіон'][:27]
plt.barh(region, years, color=['r'], label='Народжуваність')

plt.xlabel('Кількість', fontsize='small', color='black')
plt.ylabel('Область', fontsize='small', color='black')
plt.title('Народжуванність в Україні 1960 року', fontsize=15)
plt.legend()
plt.show()
```

Кругова діаграма народжуваності за 1990 рік

```python
plt.pie(
    table['1990'],
    labels=table['Регіон'],
    autopct="%.2f%%",
    pctdistance=1.35,
    labeldistance=1.5,
)

plt.show()
```

Лінійний графік за той самий рік
